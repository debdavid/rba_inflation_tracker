```python
import pandas as pd
from pathlib import Path

BRONZE = Path("../data/bronze")

print("Libraries loaded successfully")
```

```python
df_cash = pd.read_parquet(BRONZE / "rba_cash_rate.parquet")
print("Shape:", df_cash.shape)
print("\nFirst 3 rows:")
df_cash.head(3)
```

```python
print(df_cash.columns.tolist())
```

```python
df_cash_clean = df_cash[["Series ID", "FIRMMCRT"]].copy()
df_cash_clean.columns = ["date", "cash_rate"]
df_cash_clean = df_cash_clean.dropna(subset=["cash_rate"])
df_cash_clean["date"] = pd.to_datetime(df_cash_clean["date"])
df_cash_clean = df_cash_clean.sort_values("date").reset_index(drop=True)

print("Shape:", df_cash_clean.shape)
print("\nFirst 5 rows:")
df_cash_clean.head()
```

```python
print("Date range:", df_cash_clean["date"].min(), "to", df_cash_clean["date"].max())
print("Latest cash rate:", df_cash_clean["cash_rate"].iloc[-1], "%")
print("Lowest ever:", df_cash_clean["cash_rate"].min(), "%")
print("Highest ever:", df_cash_clean["cash_rate"].max(), "%")
print("\nMissing values:", df_cash_clean.isnull().sum())
```

```python
import plotly.express as px

fig = px.line(
    df_cash_clean,
    x="date",
    y="cash_rate",
    title="RBA Cash Rate Target - August 1990 to Present",
    labels={"date": "Date", "cash_rate": "Cash Rate (%)"}
)

fig.show()
```
